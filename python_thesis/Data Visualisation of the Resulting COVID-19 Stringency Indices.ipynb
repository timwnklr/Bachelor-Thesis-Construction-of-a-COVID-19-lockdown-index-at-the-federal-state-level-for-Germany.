```python
# functions for standardisation of index value
def index_intC1(cell):
    return (cell /8) * 100

def index_intC2(cell):
    return (cell /6) * 100

def index_intC6(cell):
    return (cell /2) * 100

def index_intCb(cell):
    return cell * 100

# executing the function
df_dict['C1'][federal_list] = df_dict['C1'][federal_list].apply(index_intC1)
df_dict['C2'][federal_list] = df_dict['C2'][federal_list].apply(index_intC2)
df_dict['C3'][federal_list] = df_dict['C3'][federal_list].apply(index_intCb)
df_dict['C4'][federal_list] = df_dict['C4'][federal_list].apply(index_intCb)
df_dict['C5'][federal_list] = df_dict['C5'][federal_list].apply(index_intCb)
df_dict['C6'][federal_list] = df_dict['C6'][federal_list].apply(index_intC6)
df_dict['C7'][federal_list] = df_dict['C7'][federal_list].apply(index_intCb)
```

```python
c_list = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'SI', 'SI_W']
df_ds = pd.DataFrame()

for c in c_list:
    df_ds[(str(c)) + '_mean'] = df_dict[str(c)].mean()
    df_ds[(str(c)) + '_std'] = df_dict[str(c)].std()
    df_ds[(str(c)) + '_min'] = df_dict[str(c)].min()
    df_ds[(str(c)) + '_max'] = df_dict[str(c)].max()
```

## Import & Transform & Merge COVID-19 Infection-Data

```python
c_list = ['SI_W', 'SI','C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'SI']
df_heat_max = pd.DataFrame()

for c in c_list:
    df_heat_max[(str(c))] = df_dict[str(c)].max()
   
```

```python
#Infection data
#'https://raw.githubusercontent.com/covid19-eu-zh/covid19-eu-data/master/dataset/covid-19-de.csv'
df_dict['Cov'] = pd.read_excel('/Users/timwnklr/Desktop/BA/infections.xls', index_col = None)
```

```python
df_dict['Cov']
```

```python
federal_list_ger = list(set(df_dict['Cov']['nuts_1'].values))
federal_list_ger.remove('Repatriierte')
```

```python
df_dict['Cov']['datetime'] = pd.to_datetime(df_dict['Cov']['datetime'])
df_dict['Cov']['datetime_str'] = df_dict['Cov']['datetime'].dt.strftime('%Y-%m-%d')
df_dict['Cov']['datetime'] = pd.to_datetime(df_dict['Cov']['datetime_str'])
```

```python
df_dict['Cov'][df_dict['Cov']['nuts_1'] == 'Saarland']
```

```python
for index, row in df_dict['Cov'].iterrows():
    if (row['datetime'] < date(2020, 3, 1)) or (row['datetime'] > date(2020, 11, 2)):
        df_dict['Cov'] = df_dict['Cov'].drop(index)
 
    
```

```python
list_pop = [
11103043,
13140183,
3664088,
2531071,
680130,
1852478,
6293154,
1610774,
8003421,
17925570,
4098391,
983991,
4056941,
2180684,
2910875,
2120237,]

list_fed = ["Baden-Württemberg",
"Bayern",
"Berlin",
"Brandenburg",
"Bremen",
"Hamburg",
"Hessen",
"Mecklenburg-Vorpommern",
"Niedersachsen",
"Nordrhein-Westfalen",
"Rheinland-Pfalz",
"Saarland",
"Sachsen",
"Sachsen-Anhalt",
"Schleswig-Holstein",
"Thüringen",]

dict_pop = dict(zip(list_fed, list_pop))
```

```python
for fed in federal_list_ger:
    df_dict[str(fed)] =  df_dict['Cov'][df_dict['Cov']['nuts_1'] == fed]
    df_dict[str(fed)]['new_infections'] = df_dict[str(fed)]['cases'].diff()    
    df_dict[str(fed)] = df_dict[str(fed)].drop_duplicates(subset='datetime', keep="first")
    df_dict[str(fed)] = df_dict[str(fed)].reset_index()
    
    for index, row in df_dict[str(fed)].iterrows():
        if row['new_infections'] < 0:
            df_dict[str(fed)].at[index, 'new_infections'] = 0
    
    #https://www.statistikportal.de/de/bevoelkerung/flaeche-und-bevoelkerung
    df_dict[str(fed)]['7-day'] = df_dict[str(fed)]['new_infections'].rolling(window=7).sum()/dict_pop[str(fed)]*100000
```

```python
df_dict['Bayern']
```

```python
df_dict['C1_noI'] = df_dict['C1'].reset_index()
df_dict['SI_W_oI']= df_dict['SI_W'].reset_index()

df_relpolot = pd.DataFrame()
df_relpolot['date'] = df_dict['C1_noI']['date']

for fed in federal_list_ger:
    df_relpolot[str(fed) + '_infections'] =  df_dict[str(fed)]['7-day']
    
for fed in federal_list:
    df_relpolot[str(fed) + '_SIw'] = df_dict['SI_W_oI'][str(fed)]

df_relpolot['date'] = pd.to_datetime(df_relpolot['date'])
    
        
```

## Create Data Visualization to Compare Federal States over Time

```python
def plot_inf(fed1, fed1_ger, fed2, fed2_ger, fed3, fed3_ger, fed4, fed4_ger):
    sns.set_style("whitegrid", {"grid.color": ".8", "grid.linestyle": "--"})
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, sharey=True, figsize=(10,10))
    fig.text(0.05, 0.5, '7-day Incidence Rate', ha='center', va='center', rotation='vertical', color='sandybrown')
    fig.text(0.98, 0.5, 'Weighted Stringency Index', ha='center', va='center', rotation=-90, color = 'cornflowerblue')
    plt.xticks(rotation=30,horizontalalignment='right', color='black')
    plt.subplots_adjust(hspace=0.4)
    
    ax1.plot(df_relpolot['date'], df_relpolot[str(fed1_ger)+ '_infections'], color = 'sandybrown')
    ax1.set_ylim(0, 150)
    ax1.title.set_text(fed1)
    ax1.yaxis.set_major_locator(MultipleLocator(50))
    ax1.tick_params(colors='sandybrown')
    ax1a = ax1.twinx()
    ax1a.set_ylim(0, 100)
    l1 = ax1a.plot(df_relpolot['date'], df_relpolot[str(fed1) + '_SIw'], color='cornflowerblue')
    ax1a.tick_params(colors='cornflowerblue')
    ax1a.yaxis.set_major_locator(MultipleLocator(25))
    
    
    ax2.plot(df_relpolot['date'], df_relpolot[str(fed2_ger)+ '_infections'], color = 'sandybrown')
    ax2.title.set_text(fed2)
    ax2.tick_params(colors='sandybrown')
    ax2.set_ylim(0, 150)
    ax2a = ax2.twinx()
    ax2a.set_ylim(0, 100)
    l2 = ax2a.plot(df_relpolot['date'], df_relpolot[str(fed2) + '_SIw'], color='cornflowerblue')
    ax2a.tick_params(colors='cornflowerblue')
    ax2a.yaxis.set_major_locator(MultipleLocator(25))
    
    ax3.plot(df_relpolot['date'], df_relpolot[str(fed3_ger)+ '_infections'], color = 'sandybrown')
    ax3.title.set_text(fed3)
    ax3.tick_params(colors='sandybrown')
    ax3.set_ylim(0, 150)
    ax3a = ax3.twinx()
    ax3a.set_ylim(0, 100)
    l3 = ax3a.plot(df_relpolot['date'], df_relpolot[str(fed3) + '_SIw'], color='cornflowerblue')
    ax3a.tick_params(colors='cornflowerblue')
    ax3a.yaxis.set_major_locator(MultipleLocator(25))
    
    ax4.plot(df_relpolot['date'], df_relpolot[str(fed4_ger)+ '_infections'], color = 'sandybrown')
    ax4.title.set_text(fed4)
    ax4.tick_params(axis='y', colors='sandybrown')
    ax4.set_ylim(0, 200)
    ax4a = ax4.twinx()
    ax4a.set_ylim(0, 100)
    l4 = ax4a.plot(df_relpolot['date'], df_relpolot[str(fed4) + '_SIw'], color='cornflowerblue')
    ax4a.tick_params(colors='cornflowerblue')
    ax4a.yaxis.set_major_locator(MultipleLocator(25))
    fig.savefig('data_Verlauf.eps', bbox_inches = "tight")
```
